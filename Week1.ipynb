```python
import pandas as pd

data = {'Item': ['TV', 'TV', 'TV', 'AC'],'Company': ['LG', 'VIDEOCON', 'LG', 'SONY'],'Rupees': [12000, 10000, 15000, 14000], 'USD': [700, 650, 800, 750]}

df = pd.DataFrame(data)
print(df)
```

```python
new_rows = pd.DataFrame([ {'Item': 'AC', 'Company': 'LG', 'Rupees': 18000, 'USD': 850}, {'Item': 'TV', 'Company': 'SONY', 'Rupees': 13000, 'USD': 720}])

df = pd.concat([df, new_rows], ignore_index=True)
print(df)
```

```python
max_price =df[(df['Company'] == 'LG') & (df['Item'] == 'TV')]['Rupees'].max()
print(max_price)
```

```python
total_sum =df['Rupees'].sum()
print( total_sum)
```

```python
median_usd_sony =df[df['Company'] == 'SONY']['USD'].median()
print( median_usd_sony)
```

```python
sorted_data=df.sort_values(by="Rupees")
print(sorted_data)
```

```python
import numpy as np
data = { "Name": ["Aparna", "Pankaj", "Ram", "Ramesh", "Naveen", "Krrishnav", "Bhawna"],"Degree": ["MBA", "BCA", "M.Tech", "MBA", np.nan, "BCA", "MBA"],"Score": [90.0, np.nan, 80.0, 98.0, 97.0, 78.0, 89.0]}
df3 = pd.DataFrame(data)
print(df3)
```

```python
print(df3.groupby("Degree")["Score"].max())
```

```python
df_filled = df3["Score"].fillna(76)
print(df_filled)
```

```python
df3.set_index("Name", inplace=True)
print(df3)
```

```python
df3.reset_index(inplace=True)
avg = df3.groupby(["Name", "Degree"])["Score"].mean()
print(avg)
```

```python
mba = df3[df3["Degree"] == "MBA"].shape[0]
print(mba)
```

```python
bca = df3[df3["Degree"] == "BCA"]["Score"].mode()
print( bca)
```
